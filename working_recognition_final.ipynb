## get_labeling_df : Labeling 을 해주는 함수

```python
def get_labeling_df(label_df,total_df):
    '''
    * label_df : 작업분류가 되어있는 데이터 frameStart, frameEnd, action 열을 가진다.
    * total_df : 전체데이터 파일의 frame별 데이터 전체가 들어가 있는 데이터, frame과 모든 센서값들을 가진다.
    total_df의 frame 개수만큼 notworking을 가지고 있다가 frameStart~frameEnd까지 frame의 label을 변경한다.
    '''
    label_list=['notworking']*total_df.shape[0]
    for i in range(label_df.shape[0]):
        for j in range(label_df.frameStart[i]-1,label_df.frameEnd[i]):
            label_list[j]=label_df.action[i]
    return label_list
```

## get_sec_df : 프레임을 초단위로 묶어주는 함수

---
step size가 너무 커지면 터져버리는 현상때문에 구현한 함수인데, 생각보다 성능이 많이 향상되었다.

```python
def get_sec_df(df):
    '''
    * df : labeling까지 완료한 데이터프레임
    
    30 frame을 1초로 묶는 과정이기 때문에 센서값은 물론 label(action) 또한 어떠한 처리를 통해 1초로 묶어줘야한다.
    센서값들은 30frame 간 평균을 구해 1초로 묶어준다.
    label은 더미화로 변환 후 수를 합쳐 30frame 간 빈도를 합산하고 argmax로 가장 많이나온 행동을 1초간 행동으로 분류한다.
    label까지 마친 데이터프레임에서 불필요한 feature들을 제거하고 인덱스를 초기화한 후 해당 데이터프레임을 반환한다.
    
    '''
    sec_list=[i//30 for i in df.frame] # frame을 30(1초)로 나눈 몫을 가지는 sec(ond)_list를 만든다.
    df=pd.get_dummies(df) # 머신러닝에서 기계어로 학습 시킬 수 있기하기 위해 가변수화 해준다.
    df['sec']=sec_list
    df=df.groupby('sec').mean()
    label_list=[np.argmax(i) for i in df[['label_cable','label_cap','label_notworking','label_screw']].values]
    
    for i,j in enumerate(label_list):
        if j==0:
            label_list[i]='cable'
        elif j==1:
            label_list[i]='cap'
        elif j==2:
            label_list[i]='notworking'
        else:
            label_list[i]='screw'
        
    df['label']=label_list
    df=df.drop(['label_cable','label_cap','label_notworking','label_screw'],axis=1)
    df=df.reset_index()
    return df
```

```python
sec_df=get_sec_df(total_df) # sec_df는 초 단위로 묶은 데이터이다.
```

## create_dataset : 시계열 처리에 적합한 데이터로 변환해주는 함수 

```python
def create_dataset(df, label, step=10): # 시계열 데이터는 과거의 데이터를 통해 현재의 행동을 예측 그래서 LSTM 모델을 사용
    '''
    * df : 변환할 데이터프레임
    * label : 변환할 데이터프레임의 라벨
    
    사용하지 못하는 열과 학습에 의미가 없는 3개 feature를 제거하고 label 별로 stepsize에 맞게 데이터를 묶어서 정렬한다. (그림참고)
    Example
    capping -> 10초씩 묶어서 라벨링 후 저장
    screw -> 10초씩 묶어서 라벨링 후 저장
    '''
    df=df.drop(['BEulerX','BEulerY','BEulerZ','frame'],axis=1) 
    x_data = [] # 여기 부분이 3차원 리스트로 만들어준것인가?
    y_data = []
    for i in label.unique():
        tmp_df=df.query('label=="{}"'.format(i)).reset_index()
        for j in range(tmp_df.shape[0] - step):
            x_data.append(tmp_df.drop(['label'],axis=1).iloc[j:j+step].values)
            y_data.append(i)
            
    return np.array(x_data),np.array(y_data)
```

```python
features,targets=create_dataset(sec_df,sec_df.label) # x와 y의 데이터를 받아와야 되기 때문에
```

## target이 더미화해서 들어가야 연산이 되기 때문에 더미화를 해주고 적절한 이름을 붙혀준다.

```python
targets=pd.get_dummies(pd.DataFrame(targets))
targets.columns=['cable','cap','notworking','screw']
```

```python
from sklearn.model_selection import train_test_split
```

## train,test 데이터로 나눈다. 이 때 stratify는 반드시 사용하는 것이 좋다. 

* stratify = train과 test에 각 행동이 뽑힐 확률을 맞춰준다. 
* ex) train=cable(20%),cap(20%),notworking(40%),screw(20%) 이면 테스트 역시 test=cable(20%),cap(20%),notworking(40%),screw(20%)로 구성

```python
X_train, X_test, y_train, y_test = train_test_split(features, targets, stratify=targets, random_state=11) # LSTM모델에 넣어준다.
```

```python
from tensorflow.keras.models import Sequential # 층을 구성하기 위해
from tensorflow.keras.layers import Dense # 층끼리의 연결을 위해
from tensorflow.keras.layers import LSTM # 신경망 알고리즘의 빠른 구현 / 입력촤 출력이 존재
from tensorflow.keras.layers import Dropout # 경사도(하강)의 손실을 막기 위한 규제층
import tensorflow as tf
import keras.backend.tensorflow_backend as K # Backend에서 필요한 패키지 구현
```

## LSTM MODEL (Hyper parameter)

--- 

* step : 위의 stepsize를 조정하면 성능에 영향을 미칠 수 있다.
* 층 추가 : input층과 output층은 고정이며 같은 방식으로 lstm_year3,4 같이 추가할 수 있다.
* dropout 추가 : 그레디언트 손실문제를 막기위해 규제 층인 dropout층을 추가하면 성능향상을 기대할 수는 있으나 현재 모델에서는 크게 도움이 되지 않았다.
* units : node 수, 각 층의 노드 수를 조정해 성능향상을 기대할 수 있다. (outputlayer의 노드 수는 분류할 작업의 종류 + 비작업 수로 고정) 
* activation : 활성화 함수로 relu,softmax, sigmoid가 대표적이며 다양한 함수가 있다. (!outputlayer는 분류 문제의 경우 softmax로 고정하는 것이 좋다.)
* optimizer : 옵티마이져를 통해서도 성능향상을 기대할 수 있다. (sgd, rmsprop 등)
* epochs : 반복 횟수를 조정하면 성능향상을 기대할 수 있다. (너무 높힐 시 오버피팅의 위험이 있다.)
* batch_size : 반복 횟수 1회당 몇 개의 데이터를 학습할 것인가이며 작은 크기를 주면 세세하게 학습을 하여 성능향상을 기대할 수 있지만 러닝타임이 엄청나게 증가한다.

```python
def create_model(optim, loss):
    model = Sequential() # Sequential 모델의 Object(객체)를 넣어준다. model을 통해서 구현가능

    model.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], X_train.shape[2]), name='lstm_year1'))
    #model.add(Dropout(0.2, name='dropout1')) #dropout 층

    model.add(LSTM(units = 30, activation='relu',name='lstm_year2')) 
    
    model.add(Dense(units = 4, activation='softmax', name='output_layer'))

    model.compile(optimizer = optim, loss = loss, metrics=['accuracy']) # metrics는 최종모델의 성능을 어떻게 볼래? 정확도로 볼께

    return model
```

* 단층으로
* 중간층에도 return_sequence를 True로

#### (sample, time_step,feature), stateful

* sample : 전체 길이를 좋은 step으로 잘라 나뉜 개수
* time step : 시계열 자료 길이 (200~400이 성능이 좋다고 함)
* feature : feature 수 

```python
clf = create_model('adam', 'categorical_crossentropy') # adam = optimizer
clf.summary()
```

```python
history=clf.fit(X_train, y_train, epochs = 30, batch_size = 32, shuffle=False) # return 값으로 history 정보를 반환
```

## epoch별 train 데이터의 정확도를 확인한다.

```python
import matplotlib.pyplot as plt
%matplotlib inline
history_dict = history.history 
acc_values = history_dict['accuracy'] # epcoh 마다의 검증 정확도
epochs = range(30)


plt.plot(epochs, history.history['accuracy'], 'r', label='Training accuracy')
plt.title('Training accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.gca().set_yticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_yticks()]) 
plt.legend() # label이라는 글을 추가해주기 위해서
plt.grid('off')# 눈금표시
plt.show()
```

## 학습한 적이 없는 test 데이터로 acc를 확인할 수 있는 함수이다.  (최종성능)

```python
def get_acc(pred):
    '''
    *pred : 학습한 clf 모델의 예측값을 넣어준다.
    예측값과 실제값을 비교하여 맞는 개수를 세어 전체 개수에 나눠준다.
    '''
    acc_score=0
    for i in range(pred.shape[0]):
        try:
            if np.argmax(pred[i])==np.argmax(y_test.values[i]):
                acc_score+=1
        except IndexError :
            continue
    acc=acc_score/pred.shape[0]
    print(acc_score)
    return acc
```
